```python
lisw=[]
for i in range(0,int(1e5)):
  df=data.sample(n=30, replace=True)
  lisw.append(df["WPM"].mean())
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize =(10, 7))
a=np.array(lisw)
bins=[]

bins=np.linspace(min(lisw),max(lisw),25)

up,kk,jj=ax.hist(a, bins,density=True)

# Show plot
plt.show()
```

```python
from scipy import stats
k2, p = stats.normaltest(up)
print(p)
```

```python
from scipy.stats import norm
from numpy import mean,std
mean = mean(lisw)
std = std(lisw)
min_weight = min(lisw)
max_weight = max(lisw)
values = np.linspace(min_weight, max_weight,100)
from matplotlib import pyplot
pyplot.hist(lisw, bins=25, density=True) # ,
pyplot.plot(values, norm.pdf(values, mean, std))
```

```python
lis=[]
for i in range(0,int(1e5)):
  df=data.sample(n=30, replace=True)
  lis.append(df["Accuracy"].mean())
```

```python
print(min(lis),max(lis))
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize =(10, 7))
a=np.array(lis)
bins=[]

bins=np.linspace(min(lis),max(lis),25)
u,bins,d=ax.hist(a, bins,density=True)
print(u,bins,d)

# Show plot
plt.show()
```

```python
from scipy import stats
k2, p = stats.normaltest(u)
print(p)
```

```python
from scipy.stats import norm
from numpy import mean,std
mean = mean(lis)
std = std(lis)
min_weight = min(lis)
max_weight = max(lis)
values = np.linspace(min_weight,max_weight,100)
from matplotlib import pyplot
pyplot.hist(lis, bins=30, density=True) # ,
pyplot.plot(values, norm.pdf(values, mean, std))
print(values)
```
